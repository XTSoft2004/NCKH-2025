## **Thống kê số lượng trên tập Raw_Data**

```python
import matplotlib.pyplot as plt
import pandas as pd
import os
from PIL import Image
import numpy as np
```

```python
# Define the folder paths
path_folder = r'E:\SOL\Projects\Raw_Data'

path_TheSinhVien = os.path.join(path_folder, 'Student')
path_TheThuVien = os.path.join(path_folder, 'Library')

# Count the number of images in each folder
count1 = len([f for f in os.listdir(path_TheSinhVien) if f.lower().endswith('.jpg') and os.path.isfile(os.path.join(path_TheSinhVien, f))])
count2 = len([f for f in os.listdir(path_TheThuVien) if f.lower().endswith('.jpg') and os.path.isfile(os.path.join(path_TheThuVien, f))])

print(f'Number of images in {path_TheSinhVien}:', count1)
print(f'Number of images in {path_TheThuVien}:', count2)
```

```python
import matplotlib.pyplot as plt

# Dữ liệu
folders = ['The sinh vien', 'The thu vien']
counts = [count1, count2]

# Vẽ biểu đồ
plt.bar(folders, counts, color=['blue', 'orange'])

# Thêm nhãn bên trong cột
for i, count in enumerate(counts):
    plt.text(i, count - 30, str(count), ha='center', va='center', fontsize=12, fontweight='bold', color='white')

# Thêm tiêu đề và nhãn
plt.xlabel('The sinh vien HUSC')
plt.ylabel('Số lượng ảnh')
plt.title('Biểu đồ số lượng ảnh thẻ sinh viên và thẻ thư viện')

# Hiển thị biểu đồ
plt.show()
```

## **Thống kê số lượng trên tập Augmented_Data**

```python
# Define the folder paths
path_folder = r'E:\SOL\Projects\Aug_Data'

path_TheSinhVien = os.path.join(path_folder, 'Student')
path_TheThuVien = os.path.join(path_folder, 'Library')

# Count the number of images in each folder
count3 = len([f for f in os.listdir(path_TheSinhVien) if f.lower().endswith('.jpg') and os.path.isfile(os.path.join(path_TheSinhVien, f))])
count4 = len([f for f in os.listdir(path_TheThuVien) if f.lower().endswith('.jpg') and os.path.isfile(os.path.join(path_TheThuVien, f))])

print(f'Number of images in {path_TheSinhVien}:', count3)
print(f'Number of images in {path_TheThuVien}:', count4)
```

```python
import matplotlib.pyplot as plt

# Dữ liệu
folders = ['The sinh vien', 'The thu vien']
counts = [count3, count4]

# Vẽ biểu đồ
plt.bar(folders, counts, color=['blue', 'orange'])

# Thêm nhãn bên trong cột
for i, count in enumerate(counts):
    plt.text(i, count - 30, str(count), ha='center', va='center', fontsize=12, fontweight='bold', color='white')

# Thêm tiêu đề và nhãn
plt.xlabel('The sinh vien HUSC')
plt.ylabel('Số lượng ảnh')
plt.title('Biểu đồ số lượng ảnh thẻ sinh viên và thẻ thư viện')

# Hiển thị biểu đồ
plt.show()
```

```python
items_folder1 = os.listdir(path_TheSinhVien)
items_folder2 = os.listdir(path_TheThuVien)
x = []
y = []
for item in items_folder1:
    if item.lower().endswith('.jpg'):
        path_image = os.path.join(path_TheSinhVien, item)
        image = Image.open(path_image)
        width, height = image.size
        x.append(width)
        y.append(height)
plt.scatter(x, y)

x = []
y = []
for item in items_folder2:
    if item.lower().endswith('.jpg'):
        path_image = os.path.join(path_TheThuVien, item)
        image = Image.open(path_image)
        width, height = image.size
        x.append(width)
        y.append(height)
plt.scatter(x, y)

plt.xlabel('Chiều dài')
plt.ylabel('Chiều rộng')
plt.title('Biểu đồ kích thước ảnh thẻ sinh viên và thẻ thư viện')
plt.legend(['Thẻ sinh viên', 'Thẻ thư viện'])
plt.show()
```
